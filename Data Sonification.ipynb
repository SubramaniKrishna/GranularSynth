### Dataset Analysis and Parameter Extraction

```python
# Dependencies
import numpy as np
import matplotlib.pyplot as pyp
import scipy as sp
import pandas as pd
```

```python
dataset = "eeg_data.csv"
data = pd.read_csv(dataset)
data
```

```python
# Choosing few examples to demonstrate epileptic vs non epileptic
list_people = [0,2,1,3]
```

```python
dm = data.values
# Plots of a few EEG Signals
for i in range(10):
    pyp.subplot(10,1,i+1)
    pyp.plot(dm[i][1:-1])
```

The above plots are the eeg signal values for the first 10 patients. On Visual inspection, we do not see any trends that can clearly classify them.

```python
mean = [] #Mean of the Data
var = [] #Variance of the Data
c = [] #Class of the data
for i in range(dm.shape[0]):
    mean.append(np.mean(dm[i][1:-1]))
    var.append(np.var(dm[i][1:-1]))
    if(dm[i][-1] == 5): #If patient is Epileptic
        c.append(2)
    else: #All other cases
        c.append(0)
pyp.figure()
pyp.title('Scatter Plot of Mean vs Class')
pyp.scatter(mean,c)
pyp.show()
pyp.figure()
pyp.title('Scatter Plot of Variance vs Class')
pyp.scatter(var,c)
pyp.show()
pyp.title('Scatter Plot of Mean vs Variance Classified by Class')
pyp.scatter(mean,var,c = c)
pyp.show()
```

From the $3^{rd}$ plot above, we can clearly see that the (mean,variance) pair cannot clearly differentiate(classify) epileptic vs non epileptic

```python
f_freq = []
e_sign = []
d_distri = []
for i in list_people:
    f_freq.append(50 + np.abs(dm[i][1:-1]))
    e_sign.append(np.sign(dm[i][1:-1]))
    d_distri.append(c[i])
```

```python
# Defining the socket parameters
import socket               # Import socket module
import time

s_f = socket.socket()         # Create a socket object
s_e = socket.socket()         # Create a socket object
s_d = socket.socket()         # Create a socket object

host = socket.gethostname() # Get local machine name

port_f = 12340                # Reserve a port for your service.
port_e = 12342                # Reserve a port for your service.
port_d = 12341                # Reserve a port for your service.
```

```python
# Initializing the Sockets
print('Connecting to ', host, port_f,'To  Transmit Frequency')
s_f.connect((host, port_f))
print('Connecting to ', host, port_e,'To  Transmit Envelope')
s_e.connect((host, port_e))
print('Connecting to ', host, port_d,'To  Transmit Distribution')
s_d.connect((host, port_d))
```

```python
for i in range(len(list_people)):
    print(i)
    cou = 0
    while True:
#         print(cou)
        msg_f = (str(f_freq[i][cou]) + ';').encode('utf-8')
        msg_e = (str(e_sign[i][cou]) + ';').encode('utf-8')
        msg_d = (str(d_distri[i]) + ';').encode('utf-8')
        s_f.send(msg_f)
        s_e.send(msg_e)
        s_d.send(msg_d)
        time.sleep(0.1)
        if(cou == 177):
            break
        cou = cou + 1
print("EXIT Transmission")
```
